```python
import pandas as pd

data = pd.read_html("http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

```python
type(data)
```

```python
table = data[0]
```

```python
type(table)
```

```python
table.shape
```

```python
#replace the not assigned values by NaN

import numpy as np

table.replace("Not assigned", np.nan, inplace = True)
table
```

```python
#drop the rows having Borough as Nan

table=table.dropna(subset = ["Borough"],how='all')
table
```

```python
#replace the value og neighbourhood by the value of borough if neighbourhood value is zero

table['Neighborhood'].replace(np.nan, table['Borough'], inplace=True)
```
